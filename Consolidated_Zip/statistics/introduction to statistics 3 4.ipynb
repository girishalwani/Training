```python
def bayes_th(pa,pb_given_a,pb_given_not_a):
    not_a=1-pa
    pb=pb_given_a * pa+pb_given_not_a*not_a
    pa_given_b=(pb_given_a*pa)/pb
    return pa_given_b

```

```python
pa = 0.0002
pb_given_a=0.85
pb_given_not_a=0.05

result = bayes_th(pa,pb_given_a,pb_given_not_a)
```

```python
print('P(A|B) = ',(result*100))
```
